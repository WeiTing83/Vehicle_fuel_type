```python
import pandas as pd
import matplotlib.pyplot as plot
import time
import psycopg2
from sqlalchemy import create_engine
from config import db_password
```

```python
# loading csv file
car_df=pd.read_csv("Resources/cars_summary.csv")
car_df.head(5)
#print(car_df.count)
```

```python
# dropna for each row
car_dropna_df= car_df.dropna(how="any")
# car_dropna_df.head(20)
print(car_dropna_df.count)
```

```python
car_dropna_df.head(20)
```

```python
# Seperate name and split it between manufacturer and model
car_dropna_df[['manufacturer','model']] =  car_dropna_df["name"].str.split(" ", n=1, expand=True)
car_dropna_df.head()
```

```python
# Select column
car_clean_df=car_dropna_df[["manufacturer","model","year","selling_price","fuel","seller_type","mileage"]]
car_clean_df
```

```python
car_clean_df.dtypes
```

```python
# Clean fuel types
car_clean_df['fuel'] = car_clean_df['fuel'].replace(['CNG', 'LPG'], 'Other')
```

```python
car_clean_df['mileage'] = car_clean_df['mileage'].str.extract('([0-9]+[,./]*[0-9]*)').astype(float)
```

```python
car_clean_df.head()
```

```python
car_clean_df.dtypes
```

```python
# save clean data 
car_clean_df.to_csv("Resources/car_clean.csv")
```

```python
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Used_Cars_Fuel_Type"
engine = create_engine(db_string)
```
